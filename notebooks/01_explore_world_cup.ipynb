Quick exploration of the FIFA World Cup dataset fetched via `soccerdata`.

Run `python scripts/fetch_world_cup_data.py --season 2022` first.

```python
from __future__ import annotations

import sys
from pathlib import Path

ROOT = Path.cwd().resolve()
if (ROOT / "src").exists() is False:
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))

import pandas as pd
from soccerpredict.utils import load_config
from soccerpredict.utils.paths import data_dir

cfg = load_config()
cfg
```

```python
raw = data_dir("raw")
schedule_path = raw / "FIFA_World_Cup_2022_schedule.parquet"
schedule_path, schedule_path.exists()
```

```python
schedule = pd.read_parquet(schedule_path) if schedule_path.exists() else pd.DataFrame()
schedule.head()
```
